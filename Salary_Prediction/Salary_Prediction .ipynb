```python
# Predicted values 
y_pred=model.predict(x)
print(y_pred)
```

### Checking the accuracy of model
###### it uses R^2 to find the accuracy of model
###### R^2= SSR / SST
###### where,
###### SSR( Sum square of Regression ) = sum( y_pred - y.mean() ) ** 2
###### SSE( Sum square of Error ) = sum( yi - y_pred ) ** 2
###### SST( Sum square of total ) = SSR + SSE
###### or = sum( yi - y_predict ) ** 2
###### R^2( R_square ) = SSR / SST

```python
accur=model.score(x_test,y_test)
print("Accuracy = ",round(accur*100,3)," %")
```

## Creating   LINEAR   REGRESSION   ALGORITHM

#### Importing required Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

#### Importing the training dataset

```python
df=pd.read_csv(r"C:\Users\abdul\Downloads\PROJECTS\ML PROJECTS\Salary Prediction\Salary_Train_data.csv")
df.head()
```

#### Graphically Visualizing train data in scatter plot using matplotlib.pyplot

```python
# Graphical representaion of dependant and independant

x = df['YearsExperience']
y = df['Salary']

colors = range(len(x))
plt.figure(figsize=(10, 3))
scatter = plt.scatter(x, y, c=colors, cmap='hsv')  
plt.colorbar(scatter, label='Data Point Index')
plt.title("Data Point of YearsExperience & Salary", fontsize=15)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlabel("Years Experience", fontsize=15)
plt.ylabel("Salary", fontsize=15)
plt.show()
```

#### Seperating Dependant and independant data

```python
y=df["Salary"]#.apply(lambda x : (x-df["Salary"].mean())/df["Salary"].std()) ## STANDARDIZATION
x=df["YearsExperience"]
```

#### Creating the function to find m & c We will use Gradient Descent Algo

```python
def Gradient_descent(x,y):
    global m,c
    m = 0 
    c = 0
    n = len(x)
    l_rate = 0.026
    check=0
    plt.figure(figsize=(10,3))
    while True:
        y_predct = m * x + c
        plt.plot(x,y_predct,color="g",alpha=0.6)
        MSE = (1/n) * sum ([ i**2 for i in (y-y_predct)])
        func_m = -(2/n) * sum( x * (y-y_predct) ) 
        func_c = -(2/n) * sum( y-y_predct )
        m = m - l_rate * func_m
        c = c - l_rate * func_c
        if MSE==check:
            scatter = plt.scatter(x, y, c="b",alpha=1,marker="o")  
            plt.plot(x,m*x+c,color="r")
            print(" Model Ready ")
            break
        else:
            check=MSE
            
plt.show()
```

#### Calling the function and passing x,y parameter

```python
# Function Call
Gradient_descent(x,y)
```

#### Checking values of Slope / coefficient / m and intercept / c

```python
print(f"m = {m}\nc = {c}")
```

#### Predicting the salary of given independant variabler / x

```python
y_predict=m*x+c
y_predict.head()
```

### ACCURACY OF ALGORITHM

### Checking the accuracy of model
###### it uses R^2 to find the accuracy of model
###### R^2= SSR / SST
###### where,
###### SSR( Sum square of Regression ) = sum( y_pred - y.mean() ) ** 2
###### SSE( Sum square of Error ) = sum( yi - y_pred ) ** 2
###### SST( Sum square of total ) = SSR + SSE
###### or = sum( yi - y_predict ) ** 2
###### R^2( R_square ) = SSR / SST

```python
y=np.array(y)
y_predict=np.array([m*x+c])


sum_Sqr_regrssn=sum(y_predict-y.mean())**2
SSR = sum_Sqr_regrssn.sum()
SSR
```

```python
sum_sqr_error = sum((y-y_predict)**2)
SSE = sum_sqr_error.sum()
SSE 
```

```python
sum_sqr_total=sum_Sqr_regrssn.sum()+sum_sqr_error.sum()
# sum_sqr_total=sum((y-y.mean())**2)
SST = sum_sqr_total
SST
```

```python
R_sqr=(sum_Sqr_regrssn.sum()/sum_sqr_total)
R_2 = SSR / SST
print("Accuracy = ",round(R_sqr*100,3),"%")
```
