## Importing Data Set
### -  We have 12 months, each with a different file.

```python
# ## CONCATINATING 12 FILES IN ONE SINGLE CSV FILE USING CALENDER MODULE
import calendar
months=list(calendar.month_name)
df=pd.DataFrame()
months.pop(0)
for x in months:
    x=pd.read_csv(fr"C:\Users\abdul\Downloads\projects\Salees files\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data\Sales_{x}_2019.csv")
    df=pd.concat([df,x])
```

```python
# Using Os Module to acces all the files and then merge them
import os
os.chdir(r"C:\Users\abdul\Downloads\projects\Salees files\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data")
files=os.listdir(os.getcwd())
df=pd.DataFrame()
for file in files:
    df1=pd.read_csv(f"{os.getcwd()}\{file}")
    df=pd.concat([df,df1])
```

## DATA CLEANING
### - Cleaning the null Valuse 

```python
df.info()                 # Total 186850 rows
df.isnull().sum().sum()   # Total 3270 Null rows in DataFrame
df[df.isnull()]           # Checking the null data
df.dropna(inplace=True)   # Droping Null Values
df.isnull().sum().sum()   # Total Null Values Are Zero
```

```python
# droping string values from DataFrame
df=df[df["Order Date"]!="Order Date"]
```

### Q1 what was the best month for sales ? how much was Salesed that month ?

```python
# ADDING MONTH COLUMNS
month=df["Order Date"].str[0:2]
df["Months"]=month
```

```python
# CONVERTING DATA TYPE
df["Order ID"]=df["Order ID"].astype("int64")
df["Months"]=pd.to_numeric(df["Months"])
df["Price Each"]=pd.to_numeric(df["Price Each"])
df["Quantity Ordered"]=pd.to_numeric(df["Quantity Ordered"])
```

```python
# ADDING Sales COLUMNS
df["Sales"]=df["Quantity Ordered"]*df["Price Each"]
df.head()
```

```python
df1=df.groupby("Months")[["Quantity Ordered","Sales"]].sum()
df1
## 12 th month was the best Month with total sales of 4613443.34 $
## 1 st month was thw worst Month with total sales of 1822256.73 $
```

## Graphical Represenataion

```python

plt.figure(figsize=(12, 8))
plt.bar(range(1,13),df1["Sales"])
plt.xticks(range(1,13))
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title("Months VS Sales" )
plt.show()
```

## Q3 what time should we display ads to maximize our sales ?

```python
#ADDING NEW ROW OF HOUR
df["Hour"]=df["Order Date"].str[-5:-3]
df.head()
```

```python
# GROUPING THEM BY HOUR
times=df.groupby(["Hour"])
totalsales=times["Sales"].sum()
totalsales
```

## Graphical Represenataion

```python
x_labels =list(map(str,range(0,24)))
plt.figure(figsize=(20, 8))
plt.grid(True)
plt.bar(range(0,24),totalsales)
plt.xticks(range(0,24), x_labels)
plt.xlabel('Time in Hours')
plt.ylabel('Sales')

plt.show()
# we should show maximum ads at afternoon 11 onwards and evening 18(6) onwards because most peoples are buying at that time
```

## Q5 Why that product are sold more ? is thre any relation between their price

```python
# Grouping them by Product
# Finding mean price of all products
pprice=df.groupby("Product")["Price Each"].mean()
pprice
```

## Graphical Represenataion

```python

plt.figure(figsize=(11, 5))
plt.bar(product, pprice, color='red')
plt.xticks(rotation=90)
plt.xlabel('Products')
plt.ylabel('Price')
plt.title("Products VS Price" )
plt.show
# the price of mac book and thinkpad is too high so we are not able to understand properly 
# So we will normalize the prices using logn
```

```python
# normalizing the Prices
import numpy as np
logdf = pprice.apply(lambda x: np.log(x))
logdf
```

## Graphical Represenataion

```python
# here we can understand the prices properly

plt.figure(figsize=(10, 6))
plt.bar(product, logdf, color='red')
plt.xticks(rotation=90)
plt.xlabel('Products')
plt.ylabel('Price')
plt.title("Products VS Price ( Normalize )" )
plt.show
```

## Graphical Represenataion

```python
# Relation between sales and prices


fig, ax1 = plt.subplots()
ax1.bar(product, sale, color='blue', label='Y1')
ax2 = ax1.twinx()

ax2.plot(product, pprice, color='red', label='Y2')

ax1.bar(product,sale,color='blue')
ax1.set_xticklabels(product, rotation=90)
plt.show()
```

## Q6 what product are more often sold together ?

```python
# Grouping by Order ID
# Adding colum of products sold together on basis of Order Id
df=df[df["Order ID"].duplicated(keep=False)]
df["Grouped"]=df.groupby("Order ID")["Product"].transform(lambda x : ",".join(x))

df.head()
```

```python
# Removing The Duplicates 
most=df[["Order ID","Grouped"]].drop_duplicates(subset="Order ID",keep="first")
most.head()
```

```python
# Finding the combinations of Product Sold together mostly top 10
from itertools import combinations
from collections import Counter

count=Counter()
for row in most["Grouped"]:
    row_list=row.split(",")
    count.update(Counter(combinations(row_list,2)))
final=count.most_common(10)
final
name=[]
value=[]
for x in final:
    name.append(x[0])
    value.append(x[1])
name=list(map(lambda x : str(",".join(x) ),name))
final
```

## Graphical Represenataion

```python
## 'iPhone', 'Lightning Charging Cable' is mostly sold together
plt.figure(figsize=(8, 4))
plt.bar(name, value, color='blue', label='Y1')
plt.xticks(rotation=90)
plt.show()
```
